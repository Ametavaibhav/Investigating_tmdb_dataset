### Research Question 4 :How are budget and popularity related to the revenue of the movies?

I'll be using Pearson's correlation coefficient to calculate the correlation between revenue-budget and revenue-popularity.

```python
# Function to calculate the pearson's correlation coefficient
def correlation_coeff(x,y):
    std_x=(x-x.mean())/x.std(ddof=0)
    std_y=(y-y.mean())/y.std(ddof=0)
    return(std_x*std_y).mean()
```

### Question 4a: How are budget and revenue related

```python
# Calling the function

correlation_coeff(movies_df1['budget'],movies_df1['revenue'])
```

```python
## Plotting a scatter plot between budget and revenue


plt.figure(figsize=(10,5), dpi = 100)
plt.scatter( x = movies_df1['budget'], y = movies_df1['revenue'],)
plt.xlabel('Budget->', fontsize = 13);
plt.ylabel('Revenue->', fontsize = 13);
plt.title('Scatter plot of budget and revenue of movies', fontsize = 15);
#plt.grid(True)
```

> __The value of correlation coefficient between budget and revenue of the movies is 0.83. This suggests that there is a positive relation between budget and revnue. The correlation coefficient is not perfectly positive but it suggests that movies with higher budget may tend to earn higher revenue. The scatter plot between budget and revenue also suggests a positive relationship between them.__
